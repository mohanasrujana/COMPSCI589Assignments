## Q4 Regularization Paths

```python
import numpy as np
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
import matplotlib.pyplot as plt
import warnings

# Suppress convergence warnings for cleaner output
warnings.filterwarnings('ignore', category=UserWarning)

# Load the combined .npy data
train_data = np.load("train.npy")
val_data = np.load("val.npy")
test_data = np.load("test.npy")

X_train, y_train = train_data[:, :-1], train_data[:, -1]
X_val, y_val = val_data[:, :-1], val_data[:, -1]
X_test, y_test = test_data[:, :-1], test_data[:, -1]
```

### Part (a) - Ridge Regularization Path

```python
alphas = np.logspace(-5, 4, 50)
betaarr = np.zeros((len(alphas), X_train.shape[1]))

for i, alpha in enumerate(alphas):
    ridge = Ridge(alpha=alpha)
    ridge.fit(X_train, y_train)
    betaarr[i] = ridge.coef_

plt.figure(figsize=(12, 8))
for i in range(X_train.shape[1]):
    plt.semilogx(alphas, betaarr[:, i], label=f'Weight {i+1}')

plt.xlabel('')
plt.ylabel('Weights')
plt.title('Ridge Regularization Paths for different weights vs ')
plt.legend(bbox_to_anchor=(1.05, 1)) 
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

### Part (b) - Lasso Regularization Path

```python
alphas = np.logspace(-5, 4, 50)
betaarr_lasso = np.zeros((len(alphas), X_train.shape[1]))

for i, alpha in enumerate(alphas):
    lasso = Lasso(alpha=alpha)
    lasso.fit(X_train, y_train)
    betaarr_lasso[i] = lasso.coef_
plt.figure(figsize=(12, 8))
for i in range(X_train.shape[1]):
    plt.semilogx(alphas, betaarr_lasso[:, i], label=f'Weight {i+1}')

plt.xlabel('')
plt.ylabel('Weights')
plt.title('Lasso Regularization Paths for different weights vs ')
plt.legend(bbox_to_anchor=(1.05, 1))
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
